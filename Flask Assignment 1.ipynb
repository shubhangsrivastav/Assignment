Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.
Ans:The url_for() function is used in Flask to dynamically build URLs for a specific endpoint, based on the name of the function that handles that endpoint. It takes the name of the endpoint function as an argument and returns a URL that corresponds to that endpoint.
